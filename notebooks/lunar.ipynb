```rust
use chrono::{Datelike, NaiveDate};
```

```rust
let dt = LunarDate{year: 1901, month: (4, false), day: 8};
let g = lunar_to_gregorian(dt);
```

```rust
let mut v = Vec::new();
for i in 1901..2051 {
    let dt = LunarDate{year: i, month: (8, false), day: 14};
    //let dt = dt.unwrap();
    let dt = lunar_to_gregorian(dt).unwrap_or(NaiveDate::from_ymd(1900, 1, 1));
    v.push(dt);
    let dt = LunarDate{year: i, month: (8, false), day: 15};
    //let dt = dt.unwrap();
    let dt = lunar_to_gregorian(dt).unwrap_or(NaiveDate::from_ymd(1900, 1, 1));
    v.push(dt);
    let dt = LunarDate{year: i, month: (8, false), day: 16};
    //let dt = dt.unwrap();
    let dt = lunar_to_gregorian(dt).unwrap_or(NaiveDate::from_ymd(1900, 1, 1));
    v.push(dt);
}
```

```rust
for i in v {
    println!("({:?}, {:?}, {:?}) => true,", i.year(), i.month(), i.day());
}
```

```rust
use time::{Date, PrimitiveDateTime, Time, UtcOffset, Weekday};
```

```rust
fn last_business_day(date: &Date) -> &Date {
    while !(date.weekday() == Weekday::Sunday || date.weekday() == Weekday::Saturday){
        date = date - time::Duration::days(1);
    }
    date
}
```
